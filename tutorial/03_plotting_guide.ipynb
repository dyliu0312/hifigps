# Tutorial 03: Data Visualization Guide

`hifigps` provides a range of plotting utilities built on `matplotlib`, specifically designed for visualizing astronomical sky maps, stacking results, and data residuals. 

This tutorial covers everything from basic heatmaps to specialized geometric annotations.

```python
import numpy as np
import matplotlib.pyplot as plt
from hifigps.plot import (
    plot_heatmap, plot_heatmaps, set_ticks, get_ticks_labels, make_figure, plot_axlines,
    plot_line, plot_hist, plot_sector, plot_ellipses, plot_arcs, save_plot
)
from hifigps.plot_custom import (
    plot_stack_fit_res, plot_res, plot_profile_2c, plot_profile_2c2r, plot_profile_2r,
    plot_hist_result
)
```

## 1. Generating Test Data

First, let's create a dummy 2D map with two Gaussian peaks to visualize.

```python
shape = (120, 120)
data = np.zeros(shape)
x = np.arange(shape[0])
y = np.arange(shape[1])
X, Y = np.meshgrid(x, y)
data += np.exp(-((X-40)**2 + (Y-60)**2)/100) * 100
data += np.exp(-((X-80)**2 + (Y-60)**2)/100) * 100

# Add noise
np.random.seed(42)
noisy_data = data + np.random.randn(*shape) * 5
```

## 2. Basic Heatmaps

The `plot_heatmap` function provides a quick way to show 2D arrays with sensible defaults (like percentile-based color scaling).

```python
plot_heatmap(noisy_data, title='Simulated Signal + Noise', cbar_label='Brightness [mK]');
```

## 3. Comparing Multiple Maps

Use `plot_heatmaps` to show several plots side-by-side with shared axes.

```python
plot_heatmaps([data, noisy_data], title=['Original', 'Noisy']);
```

## 4. Visualizing Stacking Results and Residuals

In research, we often want to see the original stack, the model fit, and the residuals (data - model).

```python
residual = noisy_data - data
plot_stack_fit_res([noisy_data, data, residual], 
                   vmax=[110, 110, 10], 
                   vmin=[-10, -10, -10]);
```

## 5. Line Profiles

Taking slices through the data is essential for analyzing profiles.

```python
slice_y = noisy_data[60, :]
x_axis = np.linspace(-3, 3, 120)
plot_line([x_axis], [slice_y], xlabel='Distance [deg]', ylabel='T [mK]', title='Slice through peak');
```

## 6. Geometric Annotations

Mark specific regions using geometric shapes like sectors, ellipses, and arcs.

```python
fig, ax = make_figure(figsize=(5, 5))
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

# Draw a 'smile' pattern
plot_arcs(ax, [[-2, 2], [2, 2], [0, 0]], [4, 4, 6], 
          theta1=[30, 35, 200], theta2=[145, 150, 340], ec=['k','k', 'r'])
plot_ellipses(ax, [[-2, 2], [2, 2]], 1, fc='g')
plot_ellipses(ax, [[0, 0]], 0.5, fc='b')
ax.set_title("Geometric Annotations Example");
```

## 7. Saving Your Work

Utility to save plots with high DPI and consistent formatting.

```python
save_plot(ax, "visual_guide_output.png", dpi=150)
```
