# Tutorial 04: Beam Convolution Simulation

Real radio telescopes have a finite beam size which smears the underlying sky signal. To compare simulations with observations, we must convolve our maps with the telescope beam (e.g., the FAST main beam).

This tutorial demonstrates how to use the `hifigps-convolve` CLI tool to perform this task efficiently.

```python
import os
import numpy as np
from hifigps.data import read_h5, save_h5, delete_files
from hifigps.plot import plot_heatmaps, save_plot
```

## 1. Preparing Simulation Data

First, we create a dummy 3D sky map (frequency, RA, Dec) and save it to HDF5.

```python
# Create 10 frequency slices of a 200x200 map
test_map = np.random.randn(10, 200, 200).astype(np.float32)
data_path = "simulation_input.h5"

save_h5(data_path, ["T"], [test_map])
print(f"Input data created at {data_path}")
```

## 2. Running Beam Convolution via CLI

We use the `hifigps-convolve` command. 
Usage: `hifigps-convolve [input_file] [output_dir] [n_workers]`

```python
output_dir = "./"
n_workers = 4

# Execute the CLI command using the shell
!hifigps-convolve $data_path $output_dir $n_workers
```

## 3. Loading and Comparing Results

The convolved file is saved with the suffix `_convolvedFASTbeam.hdf5`.

```python
convolved_path = "simulation_input_convolvedFASTbeam.hdf5"
convolved_data = read_h5(convolved_path, "T")

print(f"Convolved data shape: {convolved_data.shape}")

# Visualize the first frequency slice
axes = plot_heatmaps(
    [test_map[0], convolved_data[0]], 
    title=["Original Signal", "Convolved (Beam Smeared)"],
    change_ticks=False
)
save_plot(axes, "beam_smearing_comparison.png", dpi=150)
```

## 4. Cleanup

Remove the tutorial files.

```python
delete_files([data_path, convolved_path])
```
