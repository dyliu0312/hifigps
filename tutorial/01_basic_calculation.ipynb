# Tutorial 01: Basic Astronomical Calculations

This tutorial introduces the core calculation utilities provided by `hifigps`. These functions are designed to handle common astronomical conversions and system parameters for the Five-hundred-meter Aperture Spherical radio Telescope (FAST).

We use `astropy.units` extensively to ensure physical correctness.

```python
from hifigps.calculation import (
    dv2df, df2dv, dc_to_z, freq2z, z2freq, get_unit_factor,
    resolution, mainbeam, beam_solid_angle, drift_integration_time, 
    get_integration_time_FAST_HIIM, tb_sky, circular_area, SEFD, 
    flux_density_sensitivity, brightness_temperature_sensitivity,
    snu2tb, mass_to_brightness_temperature, u
)
```

## 1. Unit and Coordinate Conversions

In 21cm cosmology and HI studies, we often need to convert between velocity width (km/s), frequency width (MHz), and redshift. 

### Velocity and Frequency Widths
The 21cm line of neutral hydrogen (HI) has a rest frequency of approximately 1420.4 MHz.

```python
# Convert a velocity width of 21 km/s (typical for HI lines) to frequency width at z=0
df = dv2df(21 * u.km / u.s).to(u.MHz)
print(f"Frequency width at z=0: {df:.4f}")
```

```python
# The inverse: convert 0.1 MHz frequency width back to velocity
dv = df2dv(0.1 * u.MHz).to("km/s")
print(f"Velocity width: {dv:.4f}")
```

### Redshift and Frequency
Direct conversions between observed frequency and redshift.

```python
# Convert observed frequency (1.29 GHz) to redshift
z = freq2z(1.29 * u.GHz)
print(f"Redshift for 1.29 GHz: {z:.4f}")
```

```python
# Convert redshift (z=0.1) and a peculiar velocity (20 km/s) to observed frequency
freq = z2freq(0.1, sight_veocity=20 * u.km / u.s)
print(f"Observed frequency: {freq.to(u.GHz):.4f}")
```

## 2. FAST Telescope System Parameters

This section demonstrates how to calculate resolution and other system-level parameters for the FAST telescope.

### Resolution and Beam Parameters

```python
# Rayleigh Criterion: FAST main beam resolution at HI rest wavelength
res = resolution()
print(f"FAST resolution at rest: {res.to(u.arcmin):.2f}")

# Main beam resolution at z=0.1 (redshifted wavelength)
mb_z = mainbeam(0.1)
print(f"FAST resolution at z=0.1: {mb_z.to(u.arcmin):.2f}")
```

```python
# Beam solid angle for a 3 arcmin beam
omega = beam_solid_angle(3 * u.arcmin)
print(f"Beam solid angle: {omega.to(u.arcsec**2):.2f}")
```

### Integration Time
Calculates the effective integration time for drift scan and HI intensity mapping (HIIM) surveys.

```python
t_drift = drift_integration_time()
print(f"Drift scan integration time: {t_drift:.2f}")

t_hiim = get_integration_time_FAST_HIIM()
print(f"HIIM integration time: {t_hiim:.2f}")
```

## 3. Sensitivity and Physics Calculations

Calculate the noise levels and physical quantities like brightness temperature.

### Sensitivity Parameters

```python
# System Equivalent Flux Density (SEFD) for T_sys = 20 K
sefd = SEFD(20 * u.K)
print(f"SEFD at 20K: {sefd.to(u.Jy):.4f}")
```

```python
# Flux density sensitivity (sigma_S) for given T_sys and bandwidth
sigma_s = flux_density_sensitivity(t_sys=23.6 * u.K, delta_f=0.025 * u.MHz)
print(f"Flux density sensitivity: {sigma_s.to(u.mJy):.4f}")
```

### Mass to Brightness Temperature
A key conversion for HI stacking: converting a known HI mass to the expected observed brightness temperature ($T_b$).

```python
# Convert 10^9 solar masses at z=0.1 to brightness temperature
tb = mass_to_brightness_temperature(0.1, 1e9 * u.solMass)
print(f"Expected Tb: {tb.to(u.K):.4f}")
```
